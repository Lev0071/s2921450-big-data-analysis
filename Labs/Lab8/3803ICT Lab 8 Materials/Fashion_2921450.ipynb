The first method you will try is using neural networks. First step is the data preparation: data rescaling, label preparation.

Hint: you can use the Keras function `to_categorical`

```python
# TODO: Make the data preparation
from tensorflow.keras.utils import to_categorical

y_train_cat = to_categorical(y_train, num_classes=10)
y_test_cat = to_categorical(y_test, num_classes=10)

X_train_norm = X_train/255# noramalise to between 0-1 *1
X_test_norm = X_test/255

# TODO: reshape the image data (2D array) into input 1D array for a neural network
print(np.shape(X_train_norm))
X_train_norm = X_train_norm.reshape(X_train_norm.shape[0],np.prod(X_train_norm.shape[1:])) #*2
print(np.shape(X_train_norm))
X_test_norm = X_test_norm.reshape(X_test_norm.shape[0],np.prod(X_test_norm.shape[1:]))

#####Learning Journal#####

## *1 - 

  # RGB (Red, Green, Blue) are 8 bit each.
  # The range for each individual colour is 0-255 (as 2^8 = 256 possibilities).
  # The combination range is 256*256*256.

  # By dividing by 255, the 0-255 range can be described with a 0.0-1.0 range where 0.0 means 0 (0x00) and 1.0 means 255 (0xFF).

  # source: https://stackoverflow.com/questions/20486700/why-do-we-always-divide-rgb-values-by-255

## *2

  # X_train_norm.reshape() - pass in the dimension we want to reshape it to
  # X_train_norm.shape[0] --- First dimension is 60,000, i.e. no. of rows in training data
  # np.prod(X_train_norm.shape[1:]) --- Dimension of 28x28 taking the first and second index
```

Next step: model building with Keras. Build your neural network architecture. At first, I would recommend a light architecture: no more than 2 hidden layers, with about 10 units per layer. Put that model into a function, so that you can reuse it later.

```python
# TODO: Build your model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def my_model(input_dim):
    # Create the Sequential object
    model = Sequential()

    # Add 2 dense layers with 10 neurons each using sigmoid or relu activation
    model.add(Dense(10,input_dim=input_dim, activation="sigmoid")) # *3
    model.add(Dense(10, activation="sigmoid")) # *4
    
    # Add the output layer with one unit: the predicted result
    model.add(Dense(10, activation='softmax'))
    
    return model

my_model(X_train_norm.shape[1]).summary() # passes in the input dimension "784"

#####Learning Journal#####

## *3

  # Input Layer

  # 10 nuerons 
  # sigmoid -- a mathematical function that normalises the vlue to 0-1 range
  # But retains the relative paterns between the data points 

## *4

  # Middle layer

```

Now compile and fit your model on your training data. Since this is a multiclass classification, the loss is not `binary_crossentropy` anymore, but `categorical_crossentropy`.

```python
import os

#https://stackoverflow.com/questions/53014306/error-15-initializing-libiomp5-dylib-but-found-libiomp5-dylib-already-initial
# os.environ['KMP_DUPLICATE_LIB_OK']='True' 

# TODO: Compile and fit your model
model = my_model(X_train_norm.shape[1])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) # *5

model.fit(X_train_norm, y_train_cat, epochs=100, batch_size=128)

#####Learning Journal#####

## *5

  # loss funtion type is categorical_crossentropy -- shows how bad the model is doing
```

Once your model has been trained, compute the accuracy (and other metrics if you want) on the train and test dataset.

Be careful, Keras returns softmax output (so an array of 10 values between 0 and 1, for which the sum is equal to 1). To compute correctly the accuracy, you have to convert that array into a categorical array with zeros and a 1.

```python
# TODO: Compute the accuracy of your model
print('accuracy on train with NN:', model.evaluate(X_train_norm, y_train_cat, verbose=0)[1])
print('accuracy on test with NN:', model.evaluate(X_test_norm, y_test_cat, verbose=0)[1])
```

```python
# TODO: use any classifier you want
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier() # 7 -- Apply random forest classifier to that PCA to classify the different articles of clothing

rf.fit(X_train_pca, y_train)

print('score with RF on train', rf.score(X_train_pca, y_train))
print('score with RF on test', rf.score(X_test_pca, y_test))
```
