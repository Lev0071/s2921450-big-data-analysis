# I. Classification 

````python
# import necessary modules
from sklearn import datasets
import matplotlib.pyplot as plt

# Load the digits dataset: digits
digits = datasets.load_digits()
#  print(digits.data.shape)
````

````python
# Create data and label arrays
X = digits.data
y = digits.target

# split into training and test set
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
````

````python
#kNN = KNeighborsClassifier(n_neighbors=3)
#kNN.fit(X_train,y_train)
#kNN.score(X_test, y_test)

#for i in range(1,8):
#    kNN = KNeighborsClassifier(n_neighbors=i)
#    kNN.fit(X_train,y_train)
#    print(kNN.score(X_train, y_train))
#    print(kNN.score(X_test, y_test))



test_array = []
train_array = []

for i in range(1,9):
    kNN = KNeighborsClassifier(n_neighbors=i)
    kNN.fit(X_train,y_train)
    train_array.append(kNN.score(X_train, y_train))
    test_array.append(kNN.score(X_test, y_test))
````

## 1. k-Nearest Neighbors

````python
# plot the above into a graph
x = np.arange(1,9)
test_data = []
train_data = []
test_data.append(pd.DataFrame(test_array))
train_data.append(pd.DataFrame(train_array))
#plt.legend()
#plt.xlabel('Number of Neighbors')
#plt.ylabel('Accuracy')
#plt.show()

plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.title('k-NN: Varying Number of Neighbours')

plt.plot(x,test_array, label='Testing Accuracy')
plt.plot(x,train_array, label='Training Accuracy')

plt.legend()

plt.show()

# For the final comparison
##################
#knn_accuracy = accuracy_score(train_array, test_array)
knn_accuracy = kNN.score(X_test,y_test)
##################
````

## 2. Support Vector machine

````python
svm = SVC(kernel='linear')
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)
print(classification_report(y_test,y_pred))

# For the final comparison
##################
#from sklearn.metrics import acccuracy_score
#print(accuracy_score(y_test,y_pred)*100)
svm_accuracy = accuracy_score(y_test, y_pred)
##################
````

````python
svm_poly = SVC(kernel='poly')
svm_poly.fit(X_train,y_train)
y_pred_poly = svm_poly.predict(X_test)
print(classification_report(y_test,y_pred_poly))
````

````python
svm_sigmoid = SVC(kernel='sigmoid')
svm_sigmoid.fit(X_train,y_train)
y_pred_sigmoid = svm_poly.predict(X_test)
print(classification_report(y_test,y_pred_sigmoid))
````

````python
svm_precomputed = SVC(kernel='rbf')
svm_precomputed.fit(X_train,y_train)
y_pred_precomputed = svm_poly.predict(X_test)
print(classification_report(y_test,y_pred_precomputed))
````

# Image Compression by K-means clustering

In this exercise, you have to perform an image compression of the given picture of New York (New York.jpeg).

To do so, we want to keep not all the possible colors, but only 4.

Thus, you have to perform a clustering algorithm, with K=4, and show the image with those only 4 colors.

```
🔦 Hint:: An image is composed of pixels. Each pixel is made of 3 channels: Red, Green, Blue (RGB). Those color levels are encoded on 8 bits, meaning they are in the range [0-255]. For example, a pixel with this RGB color : (0, 0, 255) will be a very strong blue pixel.

```

````python
%matplotlib inline
# TODO: Load and review the image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image = mpimg.imread('goldcoast.jpeg')
plt.imshow(image)
plt.show()
````

````python
# TODO: Reshape the image for k-means clustering
image_reshape = image.reshape(50325, 3)
print("image: {}".format(image.shape))
print("image reshape: {}".format(image_reshape.shape))
````

````python
# TODO: Run k-means clustering on reshaped image
from sklearn.cluster import KMeans

modele_km=KMeans(n_clusters=4)
modele_km.fit(image_reshape)
y_kmeans = modele_km.predict(image_reshape)
````

````python
# TODO: Print cluster centers and construct compressed image
from matplotlib.colors import to_hex
import matplotlib.colors as mcolors
#-------
palette = modele_km.cluster_centers_[modele_km.labels_]
pallette = [(82, 129, 169), (218, 223, 224), (147, 172, 193), (168, 197, 215), (117, 170, 212)]
pallette = np.array(palette)#[np.newaxis, :, :]
#-------
compressed_image = pallette#palette[...]
print(palette)
print("compressed image: {}".format(compressed_image.shape))
````

````python
# TODO: Show the compressed image
#-------
#Dimension of the original image
rows = image.shape[0]
cols = image.shape[1]
#-------
compressed_image = compressed_image.reshape(rows, cols, 3)
plt.imshow(compressed_image/255.)
plt.show()
````
