## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python


fig2 = plt.figure()

ax1 = fig2.add_axes([0,0,1,1])
ax2 = fig2.add_axes([0.2,0.5,.2,.2])
#fig2 = fig2.add_subplot()
```

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
#Outer grapgh
ax1.plot(x,y,'b')
ax1.set_xlabel('outer x')
ax1.set_ylabel('outer y')
ax1.set_title('Outer graph')

#Inner grapgh
ax2.plot(x,y,'r')
ax2.set_xlabel('inner x')
ax2.set_ylabel('inner y')
ax2.set_title('Inner graph')

fig2

```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig3 = plt.figure()

ax1_3 = fig3.add_axes([0,0,1,1])
ax2_3 = fig3.add_axes([0.2,0.5,.4,.4])
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
x = np.arange(0,10000)
y = x*2
z = x**2

ax2_3.plot(x,y,'b')
ax2_3.set_ylim(30,50)
ax2_3.set_xlim(20,22)
ax2_3.set_xlabel('X')
ax2_3.set_ylabel('Y')
ax2_3.set_title('zoom')


ax1_3.plot(x,z,'b')
ax1_3.set_ylim(0,1000)
ax1_3.set_xlim(0,100)
ax1_3.set_xlabel('X')
ax1_3.set_ylabel('Z')


fig3
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig4, axes4 = plt.subplots(nrows=1, ncols=2)
```

```python
axes4[0]
```

```python
axes4[1]
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
axes4[0].plot(x, y, color='blue' ,linewidth=2,linestyle='dashed')
axes4[0].set_xlabel('x')
axes4[0].set_ylabel('y')
axes4[0].set_ylim(0,200)
axes4[0].set_xlim(0,100)

axes4[1].plot(x, z, color='red',linewidth=0.5)
axes4[1].set_xlabel('x')
axes4[1].set_ylabel('z')
axes4[1].set_ylim(0,10000)
axes4[1].set_xlim(0,100)


# Display the figure object    
fig4
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig5, axes5 = plt.subplots(nrows=1, ncols=2,figsize=(12,3))

axes5[0].plot(x, y, color='blue' ,linewidth=2,linestyle='dashed')
axes5[0].set_xlabel('x')
axes5[0].set_ylabel('y')
axes5[0].set_ylim(0,200)
axes5[0].set_xlim(0,100)

axes5[1].plot(x, z, color='red',linewidth=0.5)
axes5[1].set_xlabel('x')
axes5[1].set_ylabel('z')
axes5[1].set_ylim(0,10000)
axes5[1].set_xlim(0,100)
```
